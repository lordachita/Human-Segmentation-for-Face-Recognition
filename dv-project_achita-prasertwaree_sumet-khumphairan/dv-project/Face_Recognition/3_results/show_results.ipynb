```python
import numpy as np
import matplotlib.pyplot as plt
```

## Training

```python
train_loss = np.load('100-epochs/train/train_loss_list.npy')
plt.plot(train_loss)
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.savefig('100-epochs/train/train_loss_list.png')
```

```python
train_loss_selected = []
for i in range(10):
    train_loss_selected.append(train_loss[i])
```

```python
plt.plot(train_loss_selected)
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.show()
```

```python
train_top1_list = np.load('100-epochs/train/train_top1_list.npy')
plt.plot(train_top1_list)
plt.xlabel('Epoch')
plt.ylabel('Top-1 accuracy during training')
plt.savefig('100-epochs/train/train_top1_list.png')
```

```python
train_top5_list = np.load('100-epochs/train/train_top5_list.npy')
plt.plot(train_top5_list)
plt.xlabel('Epoch')
plt.ylabel('Top-5 Accuracy during training')
plt.savefig('100-epochs/train/train_top5_list.png')
```
